```python
def magiCuadrado(n):
  # Esto es puramente con fines declarativos, podría directamente llamar a la función con los valores y listo
  c= np.zeros((n, n))
  N = [x + 1 for x in list(range(n**2))]
  i= 0
  j= 0
  nm= 0
  #suma= 0

  return magiCuad(n, i, j, N, c, 0)

def diag1_val(c):
  suma= 0
  for i in range(len(c)):
    suma += c[i][i]
  return suma == sum(c[0])

def diag2_val(c):
  suma= 0
  n= len(c)
  for i in range(len(c)):
    suma += c[i][n-1-i]
  return suma == sum(c[0])

def fila_val(c, i):
  return sum(c[i])== sum(c[0])

def col_val(c, j):
  return sum(c[:, j])== sum(c[0])

def magiCuad(n, i, j, N, c, suma):

  if i==(n-1) and j==(n-1):   # si llego al último casillero
    c_l = c.copy()
    c_l[i][j] = N[0]

    if (fila_val(c_l, i) and col_val(c_l, j)): # and diag1_val(c) and diag2_val(c)):   chequeo que tanto la última fila como columna de el número mágico, y las dos diagonales
      print("Success! :D")
      print(c_l)
      return 1   # es cuadrado mágico!
    else:
      return 0   # F

  elif i==(n-1):   # si estoy por llenar una nueva columna:
    for l in range(len(N)):
      n_l = N.copy()
      c_l = c.copy()
      c_l[i][j] = n_l[l]
      n_l.pop(l)
      if col_val(c_l, j):  # si la suma de la columna me da, sigo
        #if j==0:  # si es la primera columna que lleno, ya puedo fijarme si una de las diagonales da el número mágico, si no lo hace descarto
        #  if not diag2_val(c):
        #    return 0
        suma+= magiCuad(n, i, j+1, n_l, c_l, suma)
      else:
        #print("F")
        # print(c_l)
        return 0    # la suma de la columna no me dió, descarto

  elif j==(n-1):  # si estoy por llenar otra fila
    for l in range(len(N)):
      n_l = N.copy()
      c_l = c.copy()
      c_l[i][j] = n_l[l]
      n_l.pop(l)
      if i==0 or fila_val(c_l, i):
        suma+= magiCuad(n, i+1, 0, n_l, c_l, suma)  # puede ser solución parcial, sigo sumando en la siguiente fila
      else:
        return 0  # ya sé que no es solución parcial, descarto (o sea, no la sumo ni la sigo llenando)

  else:    # si no se completó nada, sigo no más
    for l in range(len(N)):
        n_l = N.copy()
        c_l = c.copy()
        c_l[i][j] = n_l[l]
        n_l.pop(l)
        suma+= magiCuad(n, i, j+1, n_l, c_l, suma)

  return suma
```

a) (Analizando la función dada)

- Es indefinido si c < 0 porque no se puede tener cantidad de asteroides negativa.
- Es indefinido si c > j porque solo se puede comprar máximo un asteroide al día. Por lo tanto no puede ser que por ejemplo hayan pasado 3 días (j=3) y se tengan 4 asteroides al final del 3er día.

Sino es el máximo entre:

- mgn de haber comprado un asteroide en el día j
- mgn de haber vendido un asteroide en el día j
 mgn de no haber vendido ni comprado nada en el día j, es decir, mantener los mismos asteroides en j-1 y j

La recursión se dar del último día hasta el primero/los primeros. Es tipo top-down.

```python
def astroVoid(precios):
  n_dias = len(precios)

  def astroVoidAux(j, c):
    nonlocal precios
    if c < 0 or c > j+1:
      return -float('inf')

    elif j==0:
      if c==1:
        return - precios[j]
      else:
        return 0

    else:
      ganancia_comprando= astroVoidAux(j-1, c-1) - precios[j]    # resto p[j] por el coste del asteroide comprado
      ganancia_vendiendo= astroVoidAux(j-1, c+1) + precios[j]    # sumo p[j] por la ganancia del asteroide vendido
      ganancia_no_haciendo_nada = astroVoidAux(j-1, c)
      return max(ganancia_comprando, ganancia_vendiendo, ganancia_no_haciendo_nada)

  return astroVoidAux(n_dias-1, 0)
```

```python
import numpy as np

def astroVoidMemo(precios):
  n_dias = len(precios)

  # Creo la memoria
  M = np.zeros((n_dias, n_dias+1))

  # Inicializo los elementos con Nan
  M.fill(np.nan)

  def astroVoidMemoAux(j, c):
    nonlocal precios, M
    if c < 0 or c > j+1:
      return -float('inf')

    elif j==0:
      if c==1:
        return - precios[j]
      else:
        return 0

    else:
      # Si no hay valor calculado en la memoria lo calculo
      if np.isnan(M[j, c]):
        ganancia_comprando= astroVoidMemoAux(j-1, c-1) - precios[j]    # resto p[j] por el coste del asteroide comprado
        ganancia_vendiendo= astroVoidMemoAux(j-1, c+1) + precios[j]    # sumo p[j] por la ganancia del asteroide vendido
        ganancia_no_haciendo_nada = astroVoidMemoAux(j-1, c)
        M[j, c] = max(ganancia_comprando, ganancia_vendiendo, ganancia_no_haciendo_nada)

      # Devuelvo valor de memoria
      return M[j, c]

  return astroVoidMemoAux(n_dias-1, 0)
```

# 8. CortesEconómicos
---
Debemos cortar una vara de madera en varios lugares predeterminados. Sabemos que el costo de realizar un corte en una madera de longitud ℓ es ℓ (y luego de realizar ese corte quedarán 2 varas de longitudes que sumarán ℓ). Por ejemplo, si tenemos una vara de longitud 10 metros que debe ser cortada a los 2, 4 y 7 metros desde un extremo, entonces los cortes se pueden realizar, entre otras maneras, de las siguientes formas:

- Primero cortar en la posición 2, después en la 4 y después en la 7. Esta resulta en un costo de 10 + 8 + 6 = 24 porque el primer corte se hizo en una vara de longitud 10 metros, el segundo en una de 8 metros y el último en una de 6 metros.
- Cortar primero donde dice 4, después donde dice 2, y finalmente donde dice 7, con un costo
de 10 + 4 + 6 = 20, que es menor.

Queremos encontrar el mínimo costo posible de cortar una vara de longitud ℓ.

---

a) Convencerse de que el mínimo costo de cortar una vara que abarca desde i hasta j con el conjunto C de lugares de corte es j − i mas el mínimo, para todo lugar de corte c entre i y
j, de la suma entre el mínimo costo desde i hasta c y el mínimo costo desde c hasta j.

b) Escribir matemáticamente una formulación recursiva basada en a). Explicar su semántica e indicar cuáles serían los parámetros para resolver el problema.



**Res**:

\begin{align}
\ corte(j, i) = j-i~+~min\{~corte(i, c) + corte(c, j)~\}  ~~~~~~~~~~~ ~\forall~ i \leq c \leq j
\end{align}


Los parámetros son:

- j: medida del comienzo del palo
- i: medida del final del palo
(tal que al hacer j-i nos da la distancia del palo actual ℓ)
- c: medida del corte a realizar

Además tendrá que haber una variable externa C que contenga las medidas de los cortes a realizar (ci ∈ C )


---
c) Diseñar un algoritmo de PD y dar su complejidad temporal y espacial auxiliar. Comparar cómo resultaría un enfoque top-down con uno bottom-up.


```python
import numpy as np

def cortesEconomicos(C):
  n_cortes = len(C)
  largo_palo = C[-1]

  # Creo la memoria
  M = np.zeros((n_cortes, n_cortes))

  # Inicializo los elementos con Nan
  M.fill(np.nan)

  def cortesAux(i, j):
    nonlocal M

    inicio = C.index(i)
    final = C.index(j)

    # Caso base: ya no hay cortes intermedios en el segmento
    if final-inicio <= 1:
      # Devuelvo 0 porque no hay que hacer más cortes
      return 0

    else:
      # De acá voy a tomar el mínimo después
      candidatos = []

      # Para cada marca a cortar entre i y j
      for c in C[inicio+1 : final]:

        # Tomo el índice de corte en C
        corte = C.index(c)

        if np.isnan(M[inicio, corte]):
          M[inicio, corte] = cortesAux(i, c)    # cortesAux la vuelvo a llamar con los valores de i, corte y j, no con los índices

        if np.isnan(M[corte, final]):
          M[corte, final] = cortesAux(c, j)

        candidatos.append(M[inicio, corte] + M[corte, final])

      return j-i + min(candidatos)

  # Llamo y devuelvo la función recursiva:
  return cortesAux(0, largo_palo)
```

```python
# Ej: vara de longitud 10 metros que debe ser cortada a los 2, 4 y 7  metros
C = [0, 2, 4, 7, 10]

print(cortesEconomicos(C))
```

```python
# En esta versión i y j son directamente los índices.

import numpy as np

def cortesEconomicos2(C, largo_palo):

  # Le agrego los extremos: 0 al comienzo y largo_palo al final
  C.insert(0, 0)
  C.append(largo_palo)

  n = len(C)

  # Creo la memoria
  M = np.zeros((n, n))

  # Inicializo los elementos con Nan
  M.fill(np.nan)

  def cortesAux(i, j):
    nonlocal M

    # Caso base: ya no hay cortes intermedios en el segmento
    if j-i <= 1:
      # Devuelvo 0 porque no hay que hacer más cortes
      return 0

    else:
      # De acá voy a tomar el mínimo después
      candidatos = []

      # Para cada marca a cortar entre C[i] y C[j]  (for corte in C[i+1 : j])
      for c in range(i+1, j):

        if np.isnan(M[i, c]):
          M[i, c] = cortesAux(i, c)    # cortesAux la vuelvo a llamar con los valores de i, corte y j, no con los índices

        if np.isnan(M[c, j]):
          M[c, j] = cortesAux(c, j)

        candidatos.append(M[i, c] + M[c, j])

      return C[j]-C[i] + min(candidatos)

  # Llamo y devuelvo la función recursiva:

  return cortesAux(0, n-1), '\n',  M    # n-1 = índice de largo_palo
```

```python
# Bottom-up
import numpy as np

def memoria(C):

  n= len(C)

  # Creo la memoria
  M = np.zeros((n, n))

  # Los pares i y j donde j-i<=1 son el caso base se quedan en 0
  # Lleno los que no:
  for k in range(2, n):   # para cada valor de distancia posible a partir de >1
    for i in range(n):
      if i+k < n:
        j= i+k
        candidatos = []

        # Para cada marca a cortar entre C[i] y C[j]  (for corte in C[i+1 : j])
        for c in range(i+1, j):
          # Cuando paso por acá, ya debería tener los M[i, c] y M[c, j] calculados porque ya calculé las distancias menores a k
          candidatos.append(M[i, c] + M[c, j])

        M[i, j] = C[j]-C[i] + min(candidatos)

  return M

def cortesEconomicos2(C, largo_palo):

  # Le agrego los extremos: 0 al comienzo y largo_palo al final
  C.insert(0, 0)
  C.append(largo_palo)

  # Creo y lleno la memoria
  M = memoria(C)

  # Devuelvo el mínimo con i=0, j=20
  return M[0, len(C)-1]


```

Para hacerlo PD bastaría con agregar una memoria que me vaya guardando el peso acumulado desde la caja de la caja j hasta la caja i (O bien la diferencia entre el soporte máximo y el peso acumulado). En el momento que alguna de las cajas supera su soporte máximo, se descarta la solución. (Pero y si hay cajas entre j e i que decido no apilar??)

```python
def maxCajasApiladas(w, s):
    n = len(w)
    # Creamos una lista para almacenar el peso total acumulado de todas las cajas colocadas sobre cada caja
    peso_total = [w[i] for i in range(n)]
    # Creamos una lista para almacenar la cantidad máxima de cajas apiladas hasta la caja i
    max_cajas = [1] * n

    # Inicializamos el máximo con la cantidad de cajas en la pila actual
    max_actual = 1
    # Iteramos sobre todas las cajas
    for i in range(n-1,-1,-1):
        cajas=1
        # Iteramos sobre las cajas anteriores para encontrar la máxima cantidad de cajas que pueden apilarse
        for j in range(i-1,-1,-1):
            # Verificamos si la caja i puede apilarse sobre la caja j sin exceder su soporte
            if s[j] >= peso_total[i]:
                # Actualizamos el peso total acumulado de la caja i
                peso_total[i] += w[j]
                cajas+=1
                # Actualizamos el máximo si la cantidad de cajas apiladas hasta j + 1 es mayor que el máximo actual
                max_actual = max(max_cajas[i], cajas)
        # Almacenamos el máximo en la lista
        max_cajas[i] = max_actual

    # Devolvemos el máximo de la lista
    return max(max_cajas)

# Ejemplo de uso
w = [19, 7, 5, 6, 3]
s = [15, 13, 7, 8, 2]
print(maxCajasApiladas(w, s))  # Output: 4
```
